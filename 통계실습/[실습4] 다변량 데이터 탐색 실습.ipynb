```python
import pandas as pd
import warnings
warnings.simplefilter(action = 'ignore', category = FutureWarning)
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# 깨짐 방지를 위한 Font 지정
import os
if os.name =='nt' : 
    font_family = "Malgun Gothic"
else :
    font_family = "AppleGothic"

# 값이 깨지는 문제 해결을 위해 파라미터값 설정
sns.set(font=font_family, rc = {"axes.unicode_minus" : False})
```

```python
## 데이터셋 준비
# iris 붓꽃 데이터 로드
from sklearn.datasets import load_iris

iris= load_iris()
# print(iris.DESCR) -> 붓꽃 데이터의 상세정보 확인 가능
# 붓꽃 데이터 특성 확인

columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# 데이터 생성 및 준비
data = pd.DataFrame(iris.data, columns = columns)
sy = pd.Series(iris.target, dtype= 'category')
sy = sy.cat.rename_categories(iris.target_names)
data['species'] = sy

# 필터명 정의
cols =  iris.feature_names

data.head()
```

```python
# 공분산 산출
# 공분산은 수치적인 의미보다 패턴 확인에 중점
df_cov = data.cov()
df_cov
# cov = 결측값을 제외한 쌍별 공분산을 구하는 메서드
```

```python
# 공분산 히트맵 작성
# 공분산은 선형적인 관계를 측정하기때문에
# 두 변수가 비선형적으로 함께 변하는 경우는 잘 측정하지 못 한다.
ax = sns.heatmap(df_cov, annot = True)
ylim = ax.get_ylim()
ax.set_ylim(ylim[0]+0.5, ylim[1]-0.5)
plt.show()
```

```python
## 상관계수 계산
# 절대값이 3 이상이면 상관관계가 있다.

correlation_matrix = data.corr()
correlation_matrix
```

```python
# 상관계수 히트맵
ax = sns.heatmap(correlation_matrix, annot=True)
ylim = ax.get_ylim()
ax.set_ylim(ylim[0] + 0.5, ylim[1]-0.5)
plt.show()
```

```python
# 산점도 행렬 (다변량 데이터 시각화)
sns.pairplot(data, diag_kind = 'hist')
plt.show()
```

```python
# 산점도 (Scatter_plot)
data.plot(kind = 'scatter', x='sepal_length', y='petal_width')
plt.show()
```

```python
# 상관계수 유의성 검정
## 귀무가설 : sepal_length 와 petal_width 는 상관관계가 있다.
## 귀무가설 : sepal_length 와 petal_width 는 상관관계가 없다.
import scipy.stats as sp

corr_values, pval = sp.pearsonr(data['sepal_length'], data['petal_width'])

print('상관계수 : ', corr_values)
print('p_value : ', format(pval, '.55f'))

# p_value 값은 유의수준 0.05보다 작으며 따라서 기각역에 포함되어서 상관관계가 있다.
```

```python
# 교차분석하기 위한 범주형 데이터 정의
cross_data = pd.DataFrame(
                {'영양제' : ['복용','복용','복용','복용','복용','복용','복용','복용',
                         '복용','복용','복용','복용','복용','복용','복용','복용',
                         '복용','복용','복용','복용',
                        '미복용','미복용','미복용','미복용','미복용','미복용',
                         '미복용','미복용','미복용','미복용','미복용','미복용',
                         '미복용','미복용','미복용','미복용','미복용','미복용',
                         '미복용','미복용',],
                  '감기여부' : ['유','유','유','유','유','무','무','무','무','무',
                            '무','무','무','무','무','무','무','무','무','무',
                            '유','유','유','유','유','유','유','유','유','유',
                            '유','유','무','무','무','무','무','무','무','무'
                           ]
                
                }
)
cross_data.head(5)
```

```python
# 교차분석표 생성
## summary 행렬 포함한 교차표
print(pd.crosstab(cross_data['영양제'], cross_data['감기여부'], margins=True))
```

```python
## 계산하기 위한 교차표
data_crosstab = pd.crosstab(cross_data['영양제'], cross_data['감기여부'], margins=True)
```

```python
data_crosstab
```

```python
# 교차표 분석 : 카이제곱 통계량 산출
from scipy.stats import chi2_contingency

# correction=True를 적용하면 Yates' correction이 적용되어
# 검정통계량이 보수적으로 더 낮게 나옴

result = chi2_contingency(observed=data_crosstab, correction=False)
print('1. 카이제곱 통계량 :', result[0])
print('2. p-value :', result[1])
print('3. df:', result[2]) # (행의 개수 -1) * (열의 개수 -1)
print('4. 기대값 행렬:')
pd.DataFrame(result[3]).rename(index = {0:'복용', 1:'미복용', 2:'합'}, columns = {0:'유', 1:'무', 2:'합'})
```
